## Build a correlation matrix to identify relevant variables 

```python
target_value=data['cp']
chest_pain=data["target"]
np.corrcoef(target_value,chest_pain)
plt.scatter(target_value,chest_pain)
plt.show()
```

```python
target_value=data['age']
chest_pain=data["oldpeak"]
np.corrcoef(target_value,chest_pain)
plt.scatter(target_value,chest_pain)
plt.show()
```
